```python
import requests
import sqlite3

# Función para obtener el precio de compra y venta del dólar durante todo el 2023 desde el API de SUNAT
def obtener_datos_tipo_cambio_dolar():
    url = "hhttps://api.apis.net.pe/v1/tipo-cambio-sunat"
    try:
        response = requests.get(url)
        data = response.json()
        return data
    except Exception as e:
        print("Error al obtener los datos del tipo de cambio del dólar:", e)
        return None

# Función para crear la tabla 'sunat_info' y almacenar los datos en la base de datos 'base.db'
def crear_tabla_y_almacenar_datos(data):
    if data:
        try:
            conexion = sqlite3.connect('base.db')
            cursor = conexion.cursor()

            # Crear la tabla 'sunat_info' si no existe
            cursor.execute('''CREATE TABLE IF NOT EXISTS sunat_info (
                                id INTEGER PRIMARY KEY AUTOINCREMENT,
                                fecha DATE,
                                tipo_cambio_compra DECIMAL(10, 2),
                                tipo_cambio_venta DECIMAL(10, 2)
                            )''')

            # Insertar los datos en la tabla 'sunat_info'
            for registro in data:
                fecha = registro['fecha']
                tipo_cambio_compra = registro['compra']
                tipo_cambio_venta = registro['venta']
                cursor.execute('''INSERT INTO sunat_info (fecha, tipo_cambio_compra, tipo_cambio_venta) 
                                  VALUES (?, ?, ?)''', (fecha, tipo_cambio_compra, tipo_cambio_venta))
            
            conexion.commit()
            print("Datos almacenados correctamente en la tabla 'sunat_info'.")
        
        except sqlite3.Error as error:
            print("Error al trabajar con la base de datos:", error)
        finally:
            if conexion:
                conexion.close()
    else:
        print("No se pudo obtener datos del tipo de cambio del dólar.")

# Función para mostrar el contenido de la tabla 'sunat_info'
def mostrar_contenido_tabla():
    try:
        conexion = sqlite3.connect('base.db')
        cursor = conexion.cursor()

        # Consultar y mostrar el contenido de la tabla 'sunat_info'
        cursor.execute("SELECT * FROM sunat_info")
        filas = cursor.fetchall()
        if filas:
            print("Contenido de la tabla 'sunat_info':")
            for fila in filas:
                print(f"Fechas: {fila[1]}, Tipo de cambio compra: {fila[2]}, Tipo de cambio venta: {fila[3]}")
        else:
            print("La tabla 'sunat_info' está vacía.")
    
    except sqlite3.Error as error:
        print("Error al trabajar con la base de datos:", error)
    finally:
        if conexion:
            conexion.close()

def main():
    # Obtener los datos del tipo de cambio del dólar para el año 2023
    data_tipo_cambio_dolar = obtener_datos_tipo_cambio_dolar()

    # Crear la tabla y almacenar los datos en la base de datos
    crear_tabla_y_almacenar_datos(data_tipo_cambio_dolar)

    # Mostrar el contenido de la tabla 'sunat_info'
    mostrar_contenido_tabla()

if __name__ == "__main__":
    main()
```
